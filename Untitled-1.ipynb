```python
np.random.seed(42)
xi = []
n = 1000
J = 100
for _ in range(n):
    mean = np.random.uniform(0, 10, J)  # 平均ベクトルをランダムに生成
    variance = np.random.uniform(1, 5, J)  # 分散ベクトルをランダムに生成
    consumer_demand = np.abs(np.random.normal(mean, np.sqrt(variance), J))  # 正規分布に従うランダムベクトルを生成
    xi.append(consumer_demand)
```

```python
# gurobiを使って最適化問題を解く
from gurobipy import Model, GRB
# yを最小化
model = Model()
y = model.addVar()
# wの制約：wは0-1変数，w[0]>=w[1]>=w[2]>=w[3]
w = model.addVars(4, vtype=GRB.BINARY)
z = model.addVars(10, vtype=GRB.BINARY)
model.setObjective(y, GRB.MINIMIZE)
# 制約条件
model.addConstr(y >= 0)
model.addConstr(y+2*w[0]+4*w[1]+3*w[2]+5*w[3] >= 20)
model.addConstr(w[0] >= w[1])
model.addConstr(w[1] >= w[2])
model.addConstr(w[2] >= w[3])
for i in range(4):
    model.addConstr(z[i] >= w[i])
# 最適化
model.optimize()
# 結果表示
print("Optimal value:", model.objVal)
print("w:", [w[i].x for i in range(4)])
print("y:", y.x)
print("z:", [z[i].x for i in range(10)])
```
