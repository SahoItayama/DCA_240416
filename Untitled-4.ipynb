```python
import gurobipy as gp
from gurobipy import GRB
import numpy as np

# 例としての値（実際の値に置き換えてください）
w_t = np.array([4, 4, 3, 4])  # 重みの例
K = 2  # 例としてのKの値
n = len(w_t)  # w_tの長さ

# 新しいモデルを作成
model = gp.Model("maximize_weighted_sum")

# 変数を作成
s = model.addVars(n, lb=-1, ub=1, name="s")
abs_s = model.addVars(n, lb=0, name="abs_s")

# 目的関数を設定
model.setObjective(gp.quicksum(w_t[i] * s[i] for i in range(n)), GRB.MAXIMIZE)

# abs_sとsをリンクする制約を追加
for i in range(n):
    model.addConstr(abs_s[i] == gp.abs_(s[i]), f"abs_constr_{i}")

# 絶対値の合計に対する制約を追加
model.addConstr(gp.quicksum(abs_s[i] for i in range(n)) == K, "abs_sum")

# モデルを最適化
model.optimize()

# 解を出力
if model.status == GRB.OPTIMAL:
    solution = model.getAttr('X', s)
    print("最適解:")
    for i in range(n):
        print(f"s[{i}] = {solution[i]}")
else:
    print("最適解が見つかりませんでした")
```

劣勾配の計算（除外を追加）

```python
# 劣勾配の計算
def submodular_func(w_t, K, s_list):
    n = len(w_t)
    # 新しいモデルを作成
    model = gp.Model("maximize_weighted_sum")

    # 変数を作成
    s = model.addVars(n, lb=-1, ub=1, name="s")

    # 目的関数を設定
    model.setObjective(gp.quicksum(w_t[i] * s[i] for i in range(n)), GRB.MAXIMIZE)

    # 制約を追加
    # s_iの絶対値の合計がKであることを示す制約を追加
    abs_s = model.addVars(n, lb=0, name="abs_s")
    for i in range(n):
        model.addConstr(abs_s[i] == gp.abs_(s[i]))
    model.addConstr(gp.quicksum(abs_s[i] for i in range(n)) == K)

    # sのリストの中の同じ要素は選ばないようにする制約を追加
    for s_i in s_list:
        model.addConstr(gp.quicksum(s_i[i] * s[i] for i in range(n)) <= gp.quicksum(s_i[i] * s_i[i] for i in range(n)) - 1)

    # モデルを最適化
    model.optimize()

    # 解を出力
    # 解の型はnp.ndarray
    if model.status == GRB.OPTIMAL:
        solution = model.getAttr('X', s)
        s = np.array([solution[i] for i in range(n)])
        s_list.append(s)
        print("最適解s:", s) 
        print("sの長さ:", np.sum(np.abs(s)))  
        return s
        
    else:
        return None
```

凸最適化問題の実装

```python
def solve_main(T, C, M, xi, rho, s_z_t_minus_1):
    '''
    K: 供給者数
    J: 需要者数
    n: シナリオ数
    '''
    K, J = C.shape
    n = len(xi)
    # n = 3000
    # モデルの作成
    model = gp.Model("DCA")
    
    # 変数の定義
    X = model.addMVar((K, J), lb=0.0, vtype=GRB.CONTINUOUS, name="X")
    z = model.addVars(n, lb=0.0, ub=1.0, vtype=GRB.CONTINUOUS, name="z")
    
    # 目的関数の定義
    objective = gp.quicksum(C[k, j] * X[k, j] for k in range(K) for j in range(J)) + rho * gp.quicksum(z[i] for i in range(n))- gp.quicksum(z[i] * rho * s_z_t_minus_1[i] for i in range(n))
    model.setObjective(objective, sense=GRB.MINIMIZE)
    
    # 制約条件の定義
    # rhs = []
    # for i in range(n):
    #     rhs.append((1-z[i]) * xi[i])
    # start_seiyaku = time.time()
    for i in range(n):
        model.addConstr(T @ X >= (1-z[i]) * xi[i])    
    # end_seiyaku = time.time()
    # seiyaku_time = end_seiyaku-start_seiyaku
    # print("制約条件追加の処理時間：", seiyaku_time, "秒")

    # 供給制約
    for k in range(K):
        model.addConstr(gp.quicksum(X[k, j] for j in range(J)) <= M[k])
        
    # 時間上限
    model.Params.TimeLimit = 3600
    
    # 最適化の実行
    model.optimize()
    
    # 結果の表示
    # if model.status == GRB.OPTIMAL:
    #     optimal_X = np.array([[X[k, j].x for j in range(J)] for k in range(K)]) 
    #     optimal_z = np.array([z[i].x for i in range(n)])
    #     optimal_obj_value = model.objVal
    #     return optimal_X, optimal_z, optimal_obj_value, seiyaku_time
    # else:
    #     print("最適解が見つかりませんでした。")
    #     return None, None, None, seiyaku_time

    optimal_X = np.array([[X[k, j].x for k in range(K)] for j in range(J)])
    optimal_z = np.array([z[i].x for i in range(n)])
    optimal_obj_value = model.objVal

    return optimal_X, optimal_z, optimal_obj_value
```

DCAの実装

```python
def dca(T, C, M, xi, rho, epsilon, z_t_minus_1, varepsilon=0.001, max_iteration=100):
    # 初期値
    f_t_minus_1 = np.inf
    s_list = []
    start = time.time()
    iteration = 0
    # for iteration in range(max_iteration):
    while True:
        print("-----反復回数：", iteration+1, "回-----")

        # 劣勾配の計算
        s_t_minus_1 = submodular_func(z_t_minus_1, int(len(xi) * epsilon), s_list)
        # print("s_list:", s_list)
        if s_t_minus_1 is None:
            print("s_t_minus_1がNoneです。")
            break
        
        # 最適化問題を解く
        X_t, z_t, f_t = solve_main(T, C, M, xi, rho, s_t_minus_1)
        print("z_t:", z_t)

        # 収束判定
        # if np.abs(f_t_minus_1 - f_t) < varepsilon:
        #     print("解が収束しました。")
        #     break

        # 更新
        f_t_minus_1 = f_t
        z_t_minus_1 = z_t
        iteration += 1 

        if iteration == max_iteration:
            print("最大反復回数に達しました。")
            break

    print("最適解x:", X_t) 
    print("最適解z:", z_t)
    print("最適値f：", f_t)
    end = time.time()
```

```python
def process_vector(w, K):
    w = np.asarray(w, dtype=float)
    # print("w", w)
    # wを減少順に並べ替える
    sorted_indices = np.argsort(-np.abs(w))
    sorted_w = np.abs(w[sorted_indices])
    # print("sorted indices", sorted_indices)
    # print("sorted_w", sorted_w)

    # w(i)をK番目まで1、それ以外を0にする
    s = np.zeros_like(w)
    s[sorted_indices[:K]] = np.sign(w[sorted_indices[:K]])
    # print("s", s)

    return s
```

```python
def solv_pmip(T, C, M, epsilon, xi):
    '''
    K: 供給者数
    J: 需要者数
    n: シナリオ数
    '''
    
    K, J = C.shape
    n = len(xi)

    start = time.time()
    # モデルの定義
    model = gp.Model("PMIP")

    # 変数の定義
    X = model.addMVar((K, J), lb=0.0, vtype=GRB.CONTINUOUS, name="X")
    z = model.addVars(n, vtype=GRB.BINARY, name="z")

    # 目的関数の定義
    objective = gp.quicksum(C[k, j] * X[k, j] for k in range(K) for j in range(J))
    model.setObjective(objective, sense=GRB.MINIMIZE)

    # 制約条件の定義
    # rhs = []
    # for i in range(n):
    #     rhs.append()
    # print(len(rhs))
    start_seiyaku = time.time()
    for i in range(n):
        model.addConstr(T @ X >= (1-z[i]) * xi[i])    
    end_seiyaku = time.time()
    seiyaku_time = end_seiyaku-start_seiyaku
    print("制約条件追加の処理時間：", seiyaku_time)

    model.addConstr(gp.quicksum(z[i] for i in range(n)) <= n * epsilon)
    
    # 供給制約
    for k in range(K):
        model.addConstr(gp.quicksum(X[k, j] for j in range(J)) <= M[k])

    # 時間上限
    model.Params.TimeLimit = 3000
    # 最適化の実行
    model.optimize()

    # 結果の表示
    if model.status == GRB.OPTIMAL:
        optimal_X = np.array([[X[k, j].x for k in range(K)] for j in range(J)])
        optimal_z = np.array([z[i].x for i in range(n)])
        optimal_obj_value = model.objVal
        print("最適解X:", optimal_X)
        print("最適解z:", optimal_z)
        print(optimal_X.shape)
        print("最適解の目的関数値：", optimal_obj_value)
        return optimal_X, optimal_z, optimal_obj_value

        # optimal_zの値をファイルに書き込む
        # listとして保存
        # with open("optimal_z_0621.txt", mode="a") as f:
        #     f.write(str(K)+","+str(J)+"\n")
        #     f.write(str(optimal_z) + "\n")
    else:
        print("最適解は見つかりませんでした。")

    model.close()
    end = time.time()
    print("処理時間：", end-start, "秒")
```

process_vectorにs_listを追加

```python
def process_vector_s(w, K, s_list, iteration):
    w = np.asarray(w, dtype=float)
    # print("w", w)
    # wを減少順に並べ替える
    sorted_indices = np.argsort(-np.abs(w))
    sorted_w = np.abs(w[sorted_indices])
    # print("sorted indices", sorted_indices)
    # print("sorted_w", sorted_w)

    # w(i)をK番目まで1、それ以外を0にする
    s = np.zeros_like(w)
    s[sorted_indices[:K]] = np.sign(w[sorted_indices[:K]])
    # print("s", s)
    # s_listの要素とsの要素が一致していたらsを異なるものにする
    for s_i in s_list:
        if np.all(s_i == s):
            print("s_iとsが一致しています。")
            s[sorted_indices[:K-iteration]] = np.sign(w[sorted_indices[:K-iteration]])
    s_list.append(s)

    return s
```

```python
def solv_pmip_con(T, C, M, epsilon, xi):
    '''
    K: 供給者数
    J: 需要者数
    n: シナリオ数
    '''
    
    K, J = C.shape
    n = len(xi)

    start = time.time()
    # モデルの定義
    model = gp.Model("PMIP")

    # 変数の定義
    X = model.addMVar((K, J), lb=0.0, vtype=GRB.CONTINUOUS, name="X")
    z = model.addVars(n, lb=0.0, ub=1.0, vtype=GRB.CONTINUOUS, name="z")

    # 目的関数の定義
    objective = gp.quicksum(C[k, j] * X[k, j] for k in range(K) for j in range(J))
    model.setObjective(objective, sense=GRB.MINIMIZE)

    # 制約条件の定義
    # rhs = []
    # for i in range(n):
    #     rhs.append()
    # print(len(rhs))
    start_seiyaku = time.time()
    for i in range(n):
        model.addConstr(T @ X >= (1-z[i]) * xi[i])    
    end_seiyaku = time.time()
    seiyaku_time = end_seiyaku-start_seiyaku
    print("制約条件追加の処理時間：", seiyaku_time)

    model.addConstr(gp.quicksum(z[i] for i in range(n)) <= n * epsilon)
    
    # 供給制約
    for k in range(K):
        model.addConstr(gp.quicksum(X[k, j] for j in range(J)) <= M[k])

    # 時間上限
    model.Params.TimeLimit = 3000
    # 最適化の実行
    model.optimize()

    # 結果の表示
    if model.status == GRB.OPTIMAL:
        optimal_X = np.array([[X[k, j].x for k in range(K)] for j in range(J)])
        optimal_z = np.array([z[i].x for i in range(n)])
        optimal_obj_value = model.objVal
        print("最適解X:", optimal_X)
        print("最適解z:", optimal_z)
        print(optimal_X.shape)
        print("最適解の目的関数値：", optimal_obj_value)
        return optimal_X, optimal_z, optimal_obj_value

        # optimal_zの値をファイルに書き込む
        # listとして保存
        # with open("optimal_z_0621.txt", mode="a") as f:
        #     f.write(str(K)+","+str(J)+"\n")
        #     f.write(str(optimal_z) + "\n")
    else:
        print("最適解は見つかりませんでした。")

    model.close()
    end = time.time()
    print("処理時間：", end-start, "秒")
```
